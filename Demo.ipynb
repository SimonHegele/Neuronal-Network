# 1. Activation Functions

## 1.1. Step function

A map $f: \mathbb{R} \rightarrow \{0,1\}$ such that $f_t(x):= \left\{
\begin{array}{ll}
1 & x > t \\
0 & else \\
\end{array}
\right. $
for a given threshold $t \in \mathbb{R}$


```python
threshold     = 0                                       # Threshold for the step function
step_function = Step_function(threshold)                # Initializing Step_function object with chosen threshold
inputs        = linspace(-10,10,100)                    # Creates array of 100 values evenly distributed in the interval between -10 and 10
outputs       = [step_function.eval(x) for x in inputs] # Evaluation of each input

fig, axes = pyplot.subplots(figsize=(5,2.5))
axes.plot(inputs, outputs)

pyplot.show()
```

## 1.2. Generalized logistic function

```python
fig, axes = pyplot.subplots(figsize=(5,2.5))

for alpha in [1, 10, 100]:

    logistic_function = Generalized_logistic_function(alpha)          
    inputs            = linspace(-10,10,100)                    
    outputs           = [logistic_function.eval(x) for x in inputs]

    axes.plot(inputs, outputs, label=f'Alpha={alpha}')

pyplot.legend()
pyplot.show()
```
